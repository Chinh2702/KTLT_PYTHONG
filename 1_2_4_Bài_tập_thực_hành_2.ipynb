## Thực hiện trực quan hóa dữ liệu trên tập dữ liệu về bệnh tiểu đường.

## Bước 1: Chuẩn bị dữ liệu

```python
# Chuẩn bị dữ liệu
from google.colab import files
uploaded = files.upload()   # Upload file diabetes.csv

import pandas as pd

# Đọc dữ liệu
df = pd.read_csv('diabetes.csv')

# Hiển thị thông tin tổng quan
print(df.info())
print(df.head())

# Kiểm tra có giá trị thiếu không
print(df.isnull().sum())
```

## Bước 2: Trực quan hóa với thư viện Matplotlib

```python
## Case 1: Basic Biểu đồ cột trung bình Glucose theo Outcome
import matplotlib.pyplot as plt

## Case 1: Basic - Biểu đồ cột trung bình Glucose theo Outcome
avg_glucose = df.groupby('Outcome')['Glucose'].mean()

plt.figure(figsize=(6,4))
plt.bar(avg_glucose.index, avg_glucose.values, color=['skyblue', 'salmon'])
plt.xticks([0,1], ['Không tiểu đường', 'Có tiểu đường'])
plt.title('Average Glucose by Outcome (Basic)')
plt.xlabel('Tình trạng bệnh')
plt.ylabel('Mức đường huyết trung bình')
plt.show()

```

```python
## Case 2: Advanced 1: Biểu đồ hộp (Boxplot) so sánh BMI giữa hai nhóm Outcome

plt.figure(figsize=(6,5))
df.boxplot(column='BMI', by='Outcome', grid=False, patch_artist=True,
           boxprops=dict(facecolor='lightgreen', color='green'),
           medianprops=dict(color='red'))

plt.title('BMI Distribution by Outcome (Matplotlib Advanced)')
plt.suptitle('')  # Ẩn tiêu đề phụ mặc định
plt.xlabel('Outcome (0=Không bệnh, 1=Có bệnh)')
plt.ylabel('BMI')
plt.show()
```

```python
## Case 3: Advanced 2: Biểu đồ phân tán Glucose vs Age, tô màu theo Outcome

colors = ['blue' if outcome==0 else 'red' for outcome in df['Outcome']]

plt.figure(figsize=(7,5))
plt.scatter(df['Age'], df['Glucose'], c=colors, alpha=0.6)
plt.title('Glucose vs Age colored by Outcome (Advanced 2)')
plt.xlabel('Tuổi')
plt.ylabel('Đường huyết (Glucose)')
plt.legend(['Không bệnh', 'Có bệnh'])
plt.show()
```

## Bước 3: Trực quan hóa dữ liệu với thư viện Seaborn

```python
import seaborn as sns
```

```python
## Case 1: Basic: Histogram của chỉ số BMI

plt.figure(figsize=(7,5))
sns.histplot(df['BMI'], bins=20, kde=True, color='orange')
plt.title('Phân phối chỉ số BMI (Seaborn Basic)')
plt.xlabel('BMI')
plt.ylabel('Tần suất')
plt.show()
```

```python
## Case 2: Advanced 1: Boxplot Glucose theo Outcome (màu sắc + font đẹp)

plt.figure(figsize=(7,5))
sns.boxplot(data=df, x='Outcome', y='Glucose', palette='Set2')
plt.title('Glucose Distribution by Outcome (Seaborn Advanced 1)', fontsize=14)
plt.xlabel('Outcome (0=Không bệnh, 1=Có bệnh)', fontsize=12)
plt.ylabel('Glucose', fontsize=12)
plt.show()
```

```python
## Case 3: Advanced 2: Heatmap ma trận tương quan giữa các biến

plt.figure(figsize=(10,8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Diabetes Dataset (Seaborn Advanced 2)', fontsize=14)
plt.show()
```

#  Thực hiện EDA trên tập dữ liệu mua sắm tại siêu thị

## Bước 1: Chuẩn bị dữ liệu

```python
# Import các thư viện cần thiết
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu
df = pd.read_excel('Online Retail.xlsx')

# Hiển thị thông tin tổng quan
print(df.info())
print(df.head())

# Kiểm tra có giá trị thiếu không
print(df.isnull().sum())
```

## Bước 2: Trực quan hóa dữ liệu với Matplotlib

```python
## Case 1: Basic – Biểu đồ cột thể hiện số lượng sản phẩm bán ra theo quốc gia
# Tính tổng số lượng sản phẩm bán ra theo từng quốc gia
country_sales = df.groupby('Country')['Quantity'].sum().sort_values(ascending=False).head(10) # chỉ lấy top 10 quốc gia

# Vẽ biểu đồ
plt.figure(figsize=(10,5))
plt.bar(country_sales.index, country_sales.values, color='skyblue')      # tạo biểu đồ cột cơ bản
plt.title('Top 10 quốc gia có lượng bán cao nhất')
plt.xlabel('Quốc gia')
plt.ylabel('Tổng số lượng bán')
plt.xticks(rotation=45)
plt.show()
```

```python
## Case 2: Advanced 1: Biểu đồ đường thể hiện xu hướng doanh thu theo thời gian
# Chuyển cột ngày sang kiểu datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Tạo cột doanh thu
df['Revenue'] = df['Quantity'] * df['UnitPrice']

# Tính tổng doanh thu theo tháng
monthly_revenue = df.resample('ME', on='InvoiceDate')['Revenue'].sum()

# Vẽ biểu đồ đường
plt.figure(figsize=(10,5))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o', color='green')
plt.title('Xu hướng doanh thu theo thời gian')
plt.xlabel('Thời gian (theo tháng)')
plt.ylabel('Doanh thu')
plt.grid(True)
plt.show()
```

```python
## Case 3: Advanced 2: Biểu đồ tròn thể hiện tỉ trọng đơn hàng theo quốc gia
# Đếm số hóa đơn theo quốc gia
country_orders = df['Country'].value_counts().head(5)

# Vẽ biểu đồ tròn
plt.figure(figsize=(6,6))
plt.pie(country_orders.values, labels=country_orders.index, autopct='%1.1f%%', startangle=90)
plt.title('Tỷ trọng số hóa đơn theo quốc gia (Top 5)')
plt.show()
```

## Bước 3: Trực quan hóa dữ liệu với Seaborn

```python
##  Case 1: Basic: Phân phối giá sản phẩm (UnitPrice)
plt.figure(figsize=(8,5))
sns.histplot(df['UnitPrice'], bins=50, kde=True, color='orange')
plt.title('Phân phối giá sản phẩm')
plt.xlabel('Đơn giá')
plt.ylabel('Tần suất')
plt.show()
```

```python
## Case 2: Advanced 1: Mối quan hệ giữa giá và số lượng bán
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='UnitPrice', y='Quantity', alpha=0.5)
plt.title('Mối quan hệ giữa giá sản phẩm và số lượng bán')
plt.xlabel('Đơn giá')
plt.ylabel('Số lượng')
plt.show()
```

```python
## Case 3: Advanced 2: Boxplot so sánh doanh thu theo quốc gia (Top 5)
import matplotlib.pyplot as plt
import seaborn as sns

# Chọn top 5 quốc gia có nhiều giao dịch nhất
top5_countries = df['Country'].value_counts().head(5).index

plt.figure(figsize=(10,6))

# Vẽ boxplot với palette và hue hợp lệ
sns.boxplot(
    data=df[df['Country'].isin(top5_countries)],
    x='Country',
    y='Revenue',
    hue='Country',                         # thêm hue trùng với x
    legend=False,                          # ẩn legend để không bị trùng
    palette=['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#C2C2F0']
)

plt.title('Phân bố doanh thu theo quốc gia (Top 5)', fontsize=14, fontweight='bold', color='#333333')
plt.xlabel('Quốc gia', fontsize=12)
plt.ylabel('Doanh thu', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
