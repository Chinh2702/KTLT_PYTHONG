```python
# đọc dữ liệu
import pandas as pd
data=pd.read_csv('winequality-red.csv')
data.head()
```

```python
# kiểm tra thông tin và mô tả dữ liệu
print(data.shape)
print(data.info())
data.describe()
```

```python
# thống kê tần suất và kiểm tra dữ liệu
print('phân bố chất lượng rượu: ')
print(data['quality'].value_counts())
print('\ngiá trị thiếu: ')
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='quality', data=data)
plt.title('Phân bổ chất lượng rượu đỏ')
plt.show()
sns.histplot(data['alcohol'], bins=20, kde=True)
plt.title('Phân bố độ cồn')
plt.show()
sns.boxplot(x='quality', y='alcohol', data=data)
plt.title('Độ cồn theo từng mức chất lượng')
plt.show()
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Ma trận tương quan giữa các biến')
plt.show()
```
