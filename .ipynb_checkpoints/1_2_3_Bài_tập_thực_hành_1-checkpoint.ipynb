## Bước 1: Chuẩn bị dữ liệu cho trực quan hóa dữ liệu

```python
from google.colab import files
uploaded = files.upload()            # Dùng để upload file từ máy tính lên Colab
```

```python
# Import thư viện
import pandas as pd               # Dùng để xử lý dữ liệu dạng bảng (DataFrame)
import matplotlib.pyplot as plt   # Dùng để vẽ biểu đồ cơ bản
import seaborn as sns             # Dùng để vẽ biểu đồ nâng cao, đẹp hơn Matplotlib

# Đọc dữ liệu (sau khi đã upload file)
df = pd.read_csv("winequality-red.csv")

# Chọn các cột quan trọng, giữ lại các cột chính để trực quan hóa
df = df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
         'chlorides', 'free sulfur dioxide', 'total sulfur dioxide',
         'density', 'pH', 'sulphates', 'alcohol', 'quality']]

# Tạo biến TotalAcidity (tổng các loại acid) để có thêm góc nhìn so sánh
df['TotalAcidity'] = df['fixed acidity'] + df['volatile acidity'] + df['citric acid']

# Tạo sample_id để hiển thị nhãn (giống Zip code trong bài mẫu)
df = df.reset_index().rename(columns={'index': 'sample_id'})

# Sắp xếp dữ liệu theo độ cồn giảm dần
df_sorted = df.sort_values('alcohol', ascending=False).reset_index(drop=True)

#Đoạn code này giúp hiển thị từ 1 đến 10 thay vì 0 đến 9
df_sorted = df_sorted.reset_index(drop=True)
df_sorted.index = df_sorted.index + 1
df_sorted.head(10)

# Xem 10 dòng đầu tiên giúp ta chọn Top 10 rượu có nồng độ cồn cao nhất
df_sorted.head(10)

```

## Bước 2: Trực quan hóa dữ liệu với thư viện Matplotlib

```python
## Case 1: Basic: Biểu đồ cột đơn giản hiển thị 10 mẫu có độ cồn cao nhất.

plt.figure(figsize=(12,6))    # thiết lập kích thước biểu đồ
x = df_sorted['sample_id'][0:10].astype(str)
y = df_sorted['alcohol'][0:10]
plt.bar(x, y)                 #vẽ biểu đồ cột
plt.xlabel('Sample ID')
plt.ylabel('Alcohol (%)')
plt.title('Top 10 Samples with Highest Alcohol Content (Basic)')
plt.show()                    # hiển thị biểu đồ

```

```python
## Case 2: Advanced 1: Thêm màu, tiêu đề, chỉnh font, lưới ngang giúp dễ đọc hơn.
plt.figure(figsize=(12,6))
x = df_sorted['sample_id'][0:10].astype(str)
y = df_sorted['alcohol'][0:10]

plt.bar(x, y, color='skyblue', edgecolor='black')
plt.title('Top 10 Samples with Highest Alcohol Content', fontsize=15)   # thêm và chỉnh font tiêu đề
plt.xlabel('Sample ID', fontsize=12)                                    # đặt tên cho trục X
plt.ylabel('Alcohol (%)', fontsize=12)                                  # đặt tên cho trục Y
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
## Case 3: Advanced 2 Có 2 biểu đồ song song (1 về nồng độ cồn, 1 về tổng độ chua), giúp so sánh giúp ta thấy liệu rượu có nhiều cồn thì độ acid có cao không.
fig = plt.figure(figsize=(18,8))

x = df_sorted['sample_id'][0:10].astype(str)
y1 = df_sorted['alcohol'][0:10]
y2 = df_sorted['TotalAcidity'][0:10]

# Biểu đồ 1: Alcohol
plt.subplot(1,2,1)                                          # chia khung hình thành 2 phần, vẽ biểu đồ thứ nhất
plt.bar(x, y1, color='steelblue')
plt.title('Top 10 Samples - Alcohol Content', fontsize=14)
plt.xlabel('Sample ID')
plt.ylabel('Alcohol (%)')
plt.xticks(rotation=45)

# Biểu đồ 2: TotalAcidity
plt.subplot(1,2,2)                                          # vẽ biểu đồ thứ hai bên cạnh
plt.bar(x, y2, color='darkorange')                          # đổi màu
plt.title('Top 10 Samples - Total Acidity', fontsize=14)
plt.xlabel('Sample ID')
plt.ylabel('Total Acidity')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## Bước 3: Trực quan hóa dữ liệu với thư viện Seaborn

```python
## Case 1: Basic
plt.figure(figsize=(12,6))
data = df_sorted[0:10]
sns.barplot(data=data, x='sample_id', y='alcohol')    # vẽ biểu đồ cột với dữ liệu và cặp biến x, y được chọn
plt.title('Top 10 Samples with Highest Alcohol (Seaborn Basic)')
plt.xlabel('Sample ID')
plt.ylabel('Alcohol (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
## Case 2: Advanced 1 Thêm palette màu sắc, chỉnh font-size, xoay nhãn trục, đẹp hơn và trình bày chuyên nghiệp hơn.
plt.figure(figsize=(12,6))
data = df_sorted[0:10]
ax = sns.barplot(data=data, x='sample_id', y='alcohol', hue='sample_id', palette='Blues_d', legend=False)
ax.set_title('Top 10 Samples with Highest Alcohol (Seaborn Advanced)', fontsize=16)
ax.set_xlabel('Sample ID', fontsize=12)
ax.set_ylabel('Alcohol (%)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
## Case 3: Advanced 2 Hai biểu đồ Seaborn song song giúp ta xem hai góc nhìn khác nhau trong cùng một khung hình.
fig, ax = plt.subplots(1, 2, figsize=(18,8))  # chia biểu đồ thành 2 cột
data = df_sorted[0:10]

sns.set(font_scale=1.2)   # giúp chữ rõ và cân đối hơn

# Alcohol
ax1 = sns.barplot(data=data, x='sample_id', y='alcohol', hue='sample_id', palette='Blues', legend=False, ax=ax[0])
ax1.set_title('Alcohol (%) - Top 10 Samples')
ax1.set_xlabel('Sample ID')
ax1.set_ylabel('Alcohol (%)')
ax1.tick_params(axis='x', rotation=45)

# TotalAcidity
ax2 = sns.barplot(data=data, x='sample_id', y='TotalAcidity', hue='sample_id', palette='Oranges', legend=False, ax=ax[1])
ax2.set_title('Total Acidity - Top 10 Samples')
ax2.set_xlabel('Sample ID')
ax2.set_ylabel('Total Acidity')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
